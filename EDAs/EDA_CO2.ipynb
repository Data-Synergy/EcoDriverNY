## Convertir a flotantes

```python
# Data Type Conversion:
# Columns convertir a float
columns_to_convert = ['Density\n(P/Km2)', 'Agricultural Land( %)', 'Land Area(Km2)',
                      'Birth Rate', 'Co2-Emissions', 'Forested Area (%)',
                      'CPI', 'CPI Change (%)', 'Fertility Rate', 'Gasoline Price', 'GDP',
                      'Gross primary education enrollment (%)', 'Armed Forces size',
                      'Gross tertiary education enrollment (%)', 'Infant mortality',
                      'Life expectancy', 'Maternal mortality ratio', 'Minimum wage',
                      'Out of pocket health expenditure', 'Physicians per thousand',
                      'Population', 'Population: Labor force participation (%)',
                      'Tax revenue (%)', 'Total tax rate', 'Unemployment rate', 'Urban_population']

# Convertir usando la función lambda
df[columns_to_convert] = df[columns_to_convert].applymap(lambda x: float(str(x).replace(',', '').replace('$', '').replace('%', '')))
```

## Limpieza de Nulos

```python
# List of columns with missing values
columns_with_missing = df.columns[df.isnull().any()]

# Impute numerical columns with mean
numerical_columns = df.select_dtypes(include=['float64'])
numerical_columns = numerical_columns.columns[numerical_columns.isnull().any()]
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())

# Impute categorical columns with mode
categorical_columns = df.select_dtypes(include=['object'])
categorical_columns = categorical_columns.columns[categorical_columns.isnull().any()]
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])

# Verify if all missing values are handled
missing_counts = df.isnull().sum()
print(missing_counts)
```

## Top 5 Paises con mayor Emisión de CO2

```python
# Importar librerías necesarias
import matplotlib.pyplot as plt

# Ordenar el DataFrame por la columna de emisiones de CO2 en orden descendente y obtener las primeras 5 filas
top_co2_emissions_countries = df.sort_values(by='Co2-Emissions', ascending=False).head(5)

# Crear un gráfico de torta para visualizar las emisiones de CO2 de los cinco principales países
plt.figure(figsize=(8, 8))  # Ajustar el tamaño de la figura
plt.pie(top_co2_emissions_countries['Co2-Emissions'], labels=top_co2_emissions_countries['Country'], autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Top 5 Countries with Highest CO2 Emissions')
plt.axis('equal')  # Hacer que el gráfico sea un círculo
plt.show()
```

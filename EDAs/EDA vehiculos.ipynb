### EDA detallado: sobre filas de interés

```python
afv["Category"].unique()
```

```python
afvv = afv[(afv["Category"] == "Sedan/Wagon") | (afv["Category"] == "SUV") | (afv["Category"] == "Van")]
afvv.shape
```

Nos quedamos solo con los vehículos de tipo Sedan/Wagon, SUV y Van que son los propicios para la ciudad.

528 modelos distintos.

```python
afvvd = afvv.dropna(how="all")
afvvd
```

Quitamos las filas nulas, no se elimina ninguna.

```python
afvvd["Alternative Fuel Economy City"].sort_values(ascending=False)
```

Ordenamos los modelos de acuerdo a la mayor autonomía eléctrica en la ciudad. La unidad es Wh/Km

```python
afvvd_ordenado = afvvd.loc[afvvd["Alternative Fuel Economy City"].sort_values(ascending=False).index]
afvvd_ordenado
```

```python
afvvd["Fuel"].unique()
```

```python
afvvd_ordenado_elec = afvvd_ordenado[(afvvd_ordenado["Fuel"] == "Hybrid Electric") | (afvvd_ordenado["Fuel"] == "Plug-in Hybrid Electric") | (afvvd_ordenado["Fuel"] == "Electric")]
afvvd_ordenado_elec
```

Filtramos donde el combustible alternativo es eléctrico y no otros tipos de combustión como el alcohol.

```python
afvvd_ordenado_elec.columns
```

```python
afvvd_ordenado_elec_column = afvvd_ordenado_elec[['Category', 'Model', 'Model Year', 'Manufacturer', 'Fuel',
       'All-Electric Range', 'PHEV Total Range',
       'Alternative Fuel Economy City',
       'Alternative Fuel Economy Combined', 'Conventional Fuel Economy City',
       'Conventional Fuel Economy Combined']]
```
